```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings(action='ignore')
```

```python
단독다가구 = pd.read_csv('data/단독다가구_위도경도.csv')
print(단독다가구.shape)
단독다가구.head()
```

```python
단독다가구['경도'] = 단독다가구['경도'].astype(str)
단독다가구['위도'] = 단독다가구['위도'].astype(str)
단독다가구['기준'] = 단독다가구['위도'] + 단독다가구['경도']
```

```python
단독다가구.head()
```

```python
연립 = pd.read_csv('data/연립_위도경도.csv')
print(연립.shape)
연립.head()
```

```python
연립['경도'] = 연립['경도'].astype(str)
연립['위도'] = 연립['위도'].astype(str)
연립['기준'] = 연립['위도'] + 연립['경도']
```

```python
연립.head()
```

```python
오피스텔 = pd.read_csv('data/오피스텔_위도경도.csv')
print(오피스텔.shape)
오피스텔.head()
```

```python
오피스텔['경도'] = 오피스텔['경도'].astype(str)
오피스텔['위도'] = 오피스텔['위도'].astype(str)
오피스텔['기준'] = 오피스텔['위도'] + 오피스텔['경도']
```

```python
부동산 = pd.concat([단독다가구, 연립, 오피스텔]).reset_index(drop=True)
print(부동산.shape)
부동산.head()
```
