## 실내운동시설 300m 내 개수  (실내운동시설_cnt)

```python
health = pd.read_csv('data/실내운동시설.csv')
print(health.shape)
health.head()
```

```python
health.info()
```

```python
result = []
for i in tqdm(data.index):
    for j in health.index:
        start = (float(data['위도'][i]), float(data['경도'][i]))
        goal = (float(health['위도'][j]), float(health['경도'][j]))
        distance = haversine(start, goal, unit='m')
        if distance <= 300:
            result.append([data['도로명'][i], start[0], start[1], health['상호명'][j], health['도로명주소'][j], goal[0], goal[1]])
```

```python
result
```

```python
result_df = pd.DataFrame(result, columns=['도로명','위도','경도','상호명','도로명주소','실내운동시설_위도','실내운동시설_경도'])
result_df
```

```python
result_group = result_df.groupby(['도로명'])[['상호명']].count().reset_index()
result_group.columns = ['도로명','실내운동시설_cnt']
result_group
```

```python
result_group['실내운동시설_cnt'].value_counts()
```

```python
# 임시 저장
result_group.to_csv('data/연립_편의시설추가_임시(수진님).csv', index=False, encoding='utf-8')
```
