## 오전 8시

```python
rows = sample_df
```

```python
# 오전 8시
datas = []
for j in tqdm(range(len(local))):    # 0~24 : 25개 구
    
    for i in tqdm(range(len(rows))):   # 구별 25개 샘플 = 625개 
        try:
            if (__name__ == "__main__") :
                origin          = '"'+ f"{rows.iloc[i][1]},{rows.iloc[i][2]}" + '"'
                destination     = '"'+ f"{local.iloc[j][5]},{local.iloc[j][6]}" + '"'
                mode            = "transit"
                departure_time  = "1603148400"
                key             = gmaps_key

                url = "https://maps.googleapis.com/maps/api/directions/json?origin="+ origin \
                        + "&destination=" + destination \
                        + "&mode=" + mode \
                        + "&departure_time=" + departure_time\
                        + "&language=ko" \
                        + "&key=" + key   

                request         = urllib.request.Request(url)
                context         = ssl._create_unverified_context()
                response        = urllib.request.urlopen(request, context=context)
                responseText    = response.read().decode('utf-8')
                responseJson    = json.loads(responseText)

                with open("./Agent_Transit_Directions.json","w") as rltStream :
                    json.dump(responseJson,rltStream)

        
            if ( __name__ == "__main__" ) :

                wholeDict = None
                with open("./Agent_Transit_Directions.json","r") as transitJson :
                    wholeDict = dict(json.load(transitJson))

                path            = wholeDict["routes"][0]["legs"][0]
                duration_sec    = path["duration"]["value"]
                start_geo       = path["start_location"]
                end_geo         = path["end_location"]
            x = (local.iloc[j][5],local.iloc[j][6])
            y = (rows.iloc[i][1],rows.iloc[i][2])
            distance = haversine(x, y, unit = 'km')
            dur = int(duration_sec)/60    
            datas.append([local.iloc[j][2], distance, dur])
        except:
            datas.append([local.iloc[j][2], 0, 0])
            print('ha..')
        time.sleep(0.1)
```

```python
data = pd.DataFrame(datas, columns = ['area','distance','duration'])  # 속력 = 거리/시간
data
```

# 구별 평균 속도 계산
- 개별 '속력 = 거리(distance)/시간(duration)'  => mean

```python
data = pd.read_csv('data/distance(출근시간).csv', encoding='utf-8')
data
```

```python
data2 = pd.read_csv('data/distance(퇴근시간).csv', encoding='utf-8')
data2
```

```python
data_total = pd.concat([data,data2]).reset_index()
data_total.drop('index', axis=1, inplace=True)
data_total
```

```python
# 개별 '속력(speed) = 거리(distance)/시간(duration)' 
data_total['speed'] = data_total['distance']/data_total['duration']
data_total
```

```python
# 구별 평균 계산
gu_speed = data_total.groupby(['area'])[['speed']].mean()
gu_speed
```

```python
gu_speed.to_csv('data/new_speed.csv', index=False, encoding='utf-8')
```

# 속도 검증

```python
data_total
```

```python
gu_speed = gu_speed.reset_index()
gu_speed.columns = ['area','mean_speed']
gu_speed
```

```python
test = pd.merge(data_total, gu_speed, on='area', how='left')
test
```

```python
# 시간 = 거리/속력
test['new_duration'] = test['distance']/test['mean_speed']
test
```

```python
test['duration_diff'] = test['new_duration'] - test['duration']
test
```
