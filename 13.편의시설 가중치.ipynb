```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings(action='ignore')
```

```python
오피스텔 = pd.read_csv('data/officetels.csv')
print(오피스텔.shape)
오피스텔.head()
```

```python
단독다가구 = pd.read_csv('data/one_two_room.csv')
print(단독다가구.shape)
단독다가구.head()
```

```python
연립 = pd.read_csv('data/villa.csv')
print(연립.shape)
연립.head()
```

```python
total_data = pd.concat([오피스텔, 단독다가구, 연립])
print(total_data.shape)
total_data.head()
```

```python
data = pd.DataFrame(total_data['hos_num'].describe())
data
```

```python
data['mart_num'] = total_data['mart_num'].describe()
data['fast_num'] = total_data['fast_num'].describe()
data['cafe_num'] = total_data['cafe_num'].describe()
data['cul_num'] = total_data['cul_num'].describe()
data['con_num'] = total_data['con_num'].describe()
data['laun_num'] = total_data['laun_num'].describe()
data['da_num'] = total_data['da_num'].describe()
data['gym_num'] = total_data['gym_num'].describe()
```
