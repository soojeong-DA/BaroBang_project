### 좌표변환(UTM-K/GRS80TM -> WGS84) - 카카오API

```python
import requests
import json
```

```python
headers={"Authorization":"KakaoAK ***"}
```

```python
def lat_lng_test(x,y):
    url = "https://dapi.kakao.com/v2/local/geo/transcoord.json?x={}&y={}&input_coord=TM&output_coord=WGS84".format(x,y)
    result = requests.get(url, headers=headers)
    json_obj = result.json()
    return json_obj
```

```python
from tqdm.notebook import tqdm
result = []

for i in tqdm(tmp.index):
    x, y = tmp.loc[i, '좌표정보(X)'], tmp.loc[i, '좌표정보(Y)']
    json_obj = lat_lng_test(x, y)
    try:
        for document in json_obj['documents']:
            result.append([i, document['x'], document['y']])
    except:
        result.append([i,0,0])
```

```python
result
```

```python
result_df = pd.DataFrame(result, columns=['index', 'lat', 'lng'])
result_df = result_df.set_index('index')
result_df
```

```python
tmp.shape
```

```python
tmp = tmp.reset_index()
tmp
```

```python
final = pd.merge(tmp, result_df, on='index', how='left')
final.drop('index', axis=1, inplace=True)
final
```
