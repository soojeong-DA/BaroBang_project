```python
laundry1 = pd.read_csv('data/단독다가구_편의시설_세탁소.csv')
laundry1
```

# 3개 주택 유형별 - counting columns & 컬럼 영문명

```python
laundry_total = pd.read_csv('data/laundry.csv')
laundry_total
```

## 단독/다가구

```python
data = pd.read_csv('data/one_two_room.csv')
data
```

```python
data.columns
```

```python
laundry_tmp = laundry_total[laundry_total['house_kind'] == '단독/다가구'].copy()
laundry_tmp
```

```python
laundry_cnt = laundry_tmp.groupby(['criteria'])[['name']].count().reset_index()
laundry_cnt.columns = ['criteria', 'laun_num']
laundry_cnt
```

```python
daiso_tmp = daiso_total[daiso_total['house_kind'] == '단독/다가구'].copy()
daiso_tmp
```

```python
daiso_cnt = daiso_tmp.groupby(['criteria'])[['name']].count().reset_index()
daiso_cnt.columns = ['criteria', 'da_num']
daiso_cnt
```

```python
health_tmp = health_total[health_total['house_kind'] == '단독/다가구'].copy()
health_tmp
```

```python
health_cnt = health_tmp.groupby(['criteria'])[['name']].count().reset_index()
health_cnt.columns = ['criteria', 'gym_num']
health_cnt
```

```python
# laun_num 컬럼 붙이기
data2 = pd.merge(data, laundry_cnt, how='left', on='criteria')   # 688663
data2
```

```python
# da_num 컬럼 붙이기
data3 = pd.merge(data2, daiso_cnt, how='left', on='criteria')   # 688663
data3
```

```python
# he_num 컬럼 붙이기
data4 = pd.merge(data3, health_cnt, how='left', on='criteria')   # 688663
data4
```

```python
data4.isnull().sum()
```

```python
# nan값 0으로 채우기
data4['laun_num'] = data4['laun_num'].fillna(0)
data4['da_num'] = data4['da_num'].fillna(0)
data4['gym_num'] = data4['gym_num'].fillna(0)

data4.isnull().sum()
```

```python
# 00_num 컬럼 -> 정수 type으로 변경
data4['hos_num'] = data4['hos_num'].astype(int)
data4['mart_num'] = data4['mart_num'].astype(int)
data4['fast_num'] = data4['fast_num'].astype(int)
data4['cafe_num'] = data4['cafe_num'].astype(int)
data4['cul_num'] = data4['cul_num'].astype(int)
data4['con_num'] = data4['con_num'].astype(int)
data4['laun_num'] = data4['laun_num'].astype(int)
data4['da_num'] = data4['da_num'].astype(int)
data4['gym_num'] = data4['gym_num'].astype(int)
```

```python
data4.info()
```

```python
data4
```

```python
data4.to_csv('data/one_two_room.csv', index=False, encoding='utf-8')
```
