# 단독/다가구

```python
data = pd.read_csv('data/단독다가구_위도경도.csv')
print(data.shape)
data.head()
```

## 세탁소 300m 내 개수  (세탁소_cnt)

```python
laundry = pd.read_csv('data/세탁소.csv')
print(laundry.shape)
laundry.head()
```

```python
laundry.info()
```

```python
# test
start = (float(data['위도'][0]), float(data['경도'][0]))
goal = (float(laundry['위도'][0]), float(laundry['경도'][0]))
haversine(start, goal, unit='m')   # 단위 : m
```

```python
start
```

- 실제 주소간 거리가 7.8km로, haversine로 구한 거리가 잘 계산된 것을 알 수 있음

```python
result = []
for i in tqdm(data.index):
    for j in laundry.index:
        start = (float(data['위도'][i]), float(data['경도'][i]))
        goal = (float(laundry['위도'][j]), float(laundry['경도'][j]))
        distance = haversine(start, goal, unit='m')
        if distance <= 300:
            result.append([data['도로명'][i], start[0], start[1], laundry['상호명'][j], laundry['도로명주소'][j], goal[0], goal[1]])
```

```python
result
```

```python
result_df = pd.DataFrame(result, columns=['도로명','위도','경도','상호명','도로명주소','세탁소_위도','세탁소_경도'])
result_df
```

```python
result_group = result_df.groupby(['도로명'])[['상호명']].count().reset_index()
result_group.columns = ['도로명','세탁소_cnt']
result_group
```

```python
main_data = pd.merge(data, result_group, how='left', on='도로명')
main_data
```

```python
main_data.isnull().sum()
```

```python
main_data['세탁소_cnt'] = main_data['세탁소_cnt'].fillna(0)
main_data['세탁소_cnt'] = main_data['세탁소_cnt'].astype(int)
main_data
```

```python
main_data.isnull().sum()
```

```python
main_data['세탁소_cnt'].value_counts().sort_index()
```

```python
# 임시 저장
main_data.to_csv('data/단독다가구_편의시설추가_v0.csv', index=False, encoding='utf-8')
```

```python
# 편의시설
result_df['위도'] = result_df['위도'].astype(str)
result_df['경도'] = result_df['경도'].astype(str)
result_df['lat_long'] = result_df['위도'] + result_df['경도']
result_df.head()
```

```python
result_df.shape
```

```python
result_df.isnull().sum()
```

```python
result_df['주택유형'] = '단독/다가구'
```

```python
# 혹시몰라서 도로명 남겨둠
# result_df.drop(['도로명'], axis=1, inplace=True)
result_df = result_df[['주택유형', 'lat_long', '상호명','도로명주소','세탁소_위도','세탁소_경도']].copy()  #'도로명'
result_df
```

```python
# 편의시설 임시 저장
result_df.to_csv('data/단독다가구_편의시설_세탁소.csv', index=False, encoding='utf-8')
```
