## 단독다가구(one_two_room)

```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings(action='ignore')
```

```python
단독다가구 = pd.read_csv('data/one_two_room.csv', thousands=',', index_col=0)
print(단독다가구.shape)
단독다가구
```

```python
mart_group = 대형마트.groupby(['기준'])['장소명'].count().reset_index().sort_values(by='장소명')
mart_group.columns = ['criteria','mart_num']
mart_group
```

```python
fast_group = 패스트푸드.groupby(['기준'])['장소명'].count().reset_index().sort_values(by='장소명')
fast_group.columns = ['criteria','fast_num']
fast_group
```

```python
단독다가구2 = pd.merge(단독다가구, mart_group, on='criteria', how='left')
단독다가구3 = pd.merge(단독다가구2, fast_group, on='criteria', how='left')
```

```python
print(단독다가구3['mart_num'].unique())
print(단독다가구3['fast_num'].unique())
```

```python
print(단독다가구3['mart_num'].value_counts())
print(단독다가구3['fast_num'].value_counts())
```

```python
단독다가구3.head()
```

```python
단독다가구3['mart_num'] = 단독다가구3['mart_num'].fillna(0)
단독다가구3['fast_num'] = 단독다가구3['fast_num'].fillna(0)
```

```python
단독다가구3.head()
```

```python
단독다가구3.to_csv('data/one_two_room.csv', encoding='utf-8', index=False)
```
