## 필터링

```python
#월세
def solution(table, input_rent,input_deposit,input_pay,gu,job):
    input_con = input("편의시설을 입력해주세요(최대 3개) ex)편의점,패스트푸드,마트: ").split(',')
    num = len(input_con) # 편의시설의 개수
    print(job) # 직장의 위도 경도
    # 편의시설의 Column, 만점기준, Table 뽑아내기
    # Column은 편의시설 가중치 구할때, 만점기준은 편의시설 만점 기준, Table는 편의시설 상세정보를 제시할 때 필요
    # con_0, con_1, con_2 => 편의시설 개수 칼럼명
    # stand_0, stand_1, stand_2 => 각 편의시설별 만점 기준
    # table_con_0, table_con_1, table_con_2 => 편의시설의 테이블 이름
    for i in range(num): # 편의시설의 개수만큼 for문 진행 
        if input_con[i] == '병원':
            globals()['con_{}'.format(i)] = 'hos_num'
            globals()['stand_{}'.format(i)] = 5
            globals()['table_con_{}'.format(i)] = 'hospital'
        elif input_con[i] == '마트':
            globals()['con_{}'.format(i)] = 'mart_num'
            globals()['stand_{}'.format(i)] = 2
            globals()['table_con_{}'.format(i)] = 'market'
        elif input_con[i] == '패스트푸드':
            globals()['con_{}'.format(i)] = 'fast_num'
            globals()['stand_{}'.format(i)] = 2
            globals()['table_con_{}'.format(i)] = 'fastfood'
        elif input_con[i] == '카페':
            globals()['con_{}'.format(i)] = 'cafe_num'
            globals()['stand_{}'.format(i)] = 3
            globals()['table_con_{}'.format(i)] = 'cafe'
        elif input_con[i] == '문화시설':
            globals()['con_{}'.format(i)] = 'cul_num'
            globals()['stand_{}'.format(i)] = 5
            globals()['table_con_{}'.format(i)] = 'culture'
        elif input_con[i] == '편의점':
            globals()['con_{}'.format(i)] = 'con_num'
            globals()['stand_{}'.format(i)] = 10
            globals()['table_con_{}'.format(i)] = 'convenience'
        elif input_con[i] == '세탁소':
            globals()['con_{}'.format(i)] = 'laun_num'
            globals()['stand_{}'.format(i)] = 7
            globals()['table_con_{}'.format(i)] = 'laundry'
        elif input_con[i] == '다이소':
            globals()['con_{}'.format(i)] = 'da_num'
            globals()['stand_{}'.format(i)] = 2
            globals()['table_con_{}'.format(i)] = 'daiso'
        elif input_con[i] == '실내운동시설':
            globals()['con_{}'.format(i)] = 'gym_num'
            globals()['stand_{}'.format(i)] = 4
            globals()['table_con_{}'.format(i)] = 'health'
    
    # 만약 편의시설을 3개 입력했을때
    if num == 3:
    # FIRST 부분
    # 1. address, rent, deposit, pay,latitude, longitude, area, criteria, recent, station_ar, transfer 가져옴
    # 2. {con_0},{con_1},{con_2} 가져옴(메인데이터 편의시설 칼럼명)
    # 3. con_0이 만점 기준 stand_0 보다 크다면 30점 부여하고 아니면 (30 / {stand_0} * {con_0}) 만큼 부여 => con0으로 저장
    # 4. station_ar가 2면 50점, 1면 30점, 0이면 10점 부텨
    # 5. transfer가 1이면 50, 아니면 10점 부여
    # 6. ST_Distance : SQL에서 두 저점간의 거리(각 매물위치와 직장과의 직선거리)를 계산함 
        # - m단위로 나오므로 1000으로 나누어져야 km단위가 됨
    # 7. 지정한 {table}명에서 데이터 가지고옴
    
    # 밑에부분
    # 1. address, rent, deposit, pay, latitude, longitude, criteria 등을 가져옴
    # 2. km/{gu} -> 직장과의 거리와 직장이 속한 구의 대중교통 평균속도를 나누어줌(시간 = 거리/속력)
        # -> 즉 직장과 매물과의 예상 대중교통 소요시간임
    # 3. km/{gu}가 10분보다 작으면 만점, 90분까지는 가중치계산, 90분 이상은 0점
    # 4. score 계산 : ((con0 + con1 + con2 + 10)/10 + (st1 + st2)/20 + time/10) as score
        # -> 편의서설 만점 10점, 지하철 만점 5점, 대중교통 소요시간 만점 10점 총 25점 만점(편의시설 없으면 15점 만점)
        # 편의시설 3개 선택시 각각 3점 만점 이며 1점은 기본 점수 부여
        # 편의시설 2개 선택시 각각 5점 만점 부여
        # 편의서설 1개 선택시 10점 부여
        # 편의시설 선택을 안할시에 점수부여하지 않음
    # 5. Score기준 상위 10개를 출력
        sql = (f"WITH FIRST AS( \
                    SELECT * \
                    FROM \
                        (SELECT address, rent, size::float, deposit::integer, pay::integer, latitude::float, \
                        longitude::float, area, criteria, recent,station_ar, transfer,\
                       {con_0},{con_1},{con_2}, \
                       (CASE WHEN {con_0} >= {stand_0} THEN 30 ELSE (30 / {stand_0} * {con_0})::float END) AS con0, \
                       (CASE WHEN {con_1} >= {stand_1} THEN 30 ELSE (30 / {stand_1} * {con_1})::float END) AS con1, \
                       (CASE WHEN {con_2} >= {stand_2} THEN 30 ELSE (30 / {stand_2} * {con_2})::float END) AS con2,  \
                       (CASE WHEN station_ar = 2 THEN 50 \
                             WHEN station_ar = 1 THEN 30 ELSE 10 END) as st1,\
                       (CASE WHEN transfer = 1 THEN 50 ELSE 10 END) as st2, \
                       ST_Distance(geography(ST_MakePoint(longitude, latitude)),  \
                          geography(ST_MakePoint('{job[1]}', '{job[0]}')))/1000 as km \
                        FROM {table} \
                        WHERE rent='{input_rent}' and deposit <= '{input_deposit}' \
                        and pay <= '{input_pay}' and recent = 1) as r) \
                    SELECT address, rent, size, deposit, pay, latitude, longitude, criteria, km/{gu} as distance,\
                    ((con0 + con1 + con2 + 10)/10 + (st1 + st2)/20 + time/10) as score \
                    From (SELECT *, \
                            (CASE WHEN km/{gu} <= 10 THEN 100 \
                                  WHEN km/{gu} between 10 and 90 THEN (100 - km/{gu}) ELSE 0 END) as time\
                          FROM FIRST) as k \
                    ORDER BY score desc LIMIT 10;")   
    # 편의시설 2개일때
    elif num == 2:
        sql = (f"WITH FIRST AS( \
                    SELECT * \
                    FROM \
                        (SELECT address, rent, size::float, deposit::integer, pay::integer, latitude::float, \
                        longitude::float, area, criteria, recent,station_ar, transfer,\
                       {con_0},{con_1},\
                       (CASE WHEN {con_0} >= {stand_0} THEN 50 ELSE (50 / {stand_0} * {con_0})::float END) AS con0, \
                       (CASE WHEN {con_1} >= {stand_1} THEN 50 ELSE (50 / {stand_1} * {con_1})::float END) AS con1, \
                       (CASE WHEN station_ar = 2 THEN 50 \
                             WHEN station_ar = 1 THEN 30 ELSE 10 END) as st1,\
                       (CASE WHEN transfer = 1 THEN 50 ELSE 10 END) as st2, \
                       ST_Distance(geography(ST_MakePoint(longitude, latitude)),  \
                          geography(ST_MakePoint('{job[1]}', '{job[0]}')))/1000 as km \
                        FROM {table} \
                        WHERE rent='{input_rent}' and deposit <= '{input_deposit}' \
                        and pay <= '{input_pay}' and recent = 1) as r) \
                    SELECT address, rent, size, deposit, pay, latitude, longitude, criteria, km/{gu} as distance,\
                    ((con0 + con1)/10 + (st1 + st2)/20 + time/10) as score \
                    From (SELECT *, \
                            (CASE WHEN km/{gu} <= 10 THEN 100 \
                                  WHEN km/{gu} between 10 and 90 THEN (100 - km/{gu}) ELSE 0 END) as time\
                          FROM FIRST) as k \
                    ORDER BY score desc LIMIT 10;")   
    # 편의시설 1개일때
    elif num == 1:
        sql = (f"WITH FIRST AS( \
                    SELECT * \
                    FROM \
                        (SELECT address, rent, size::float, deposit::integer, pay::integer, latitude::float, \
                        longitude::float, area, criteria, recent,station_ar, transfer,\
                       {con_0},\
                       (CASE WHEN {con_0} >= {stand_0} THEN 100 ELSE (100 / {stand_0} * {con_0})::float END) AS con0, \
                       (CASE WHEN station_ar = 2 THEN 50 \
                             WHEN station_ar = 1 THEN 30 ELSE 10 END) as st1,\
                       (CASE WHEN transfer = 1 THEN 50 ELSE 10 END) as st2, \
                       ST_Distance(geography(ST_MakePoint(longitude, latitude)),  \
                          geography(ST_MakePoint('{job[1]}', '{job[0]}')))/1000 as km \
                        FROM {table} \
                        WHERE rent='{input_rent}' and deposit <= '{input_deposit}' \
                        and pay <= '{input_pay}' and recent = 1) as r) \
                    SELECT address, rent, size, deposit, pay, latitude, longitude, criteria, km/{gu} as distance,\
                    (con0 /10 + (st1 + st2)/20 + time/10) as score \
                    From (SELECT *, \
                            (CASE WHEN km/{gu} <= 10 THEN 100 \
                                  WHEN km/{gu} between 10 and 90 THEN (100 - km/{gu}) ELSE 0 END) as time\
                          FROM FIRST) as k \
                    ORDER BY score desc LIMIT 10;")                
    # 편의시설 안 골랐을 때
    else:
        sql = (f"WITH FIRST AS( \
                    SELECT * \
                    FROM \
                        (SELECT address, rent, size::float, deposit::integer, pay::integer, latitude::float, \
                        longitude::float, area, criteria, recent,station_ar, transfer,\
                       (CASE WHEN station_ar = 2 THEN 50 \
                             WHEN station_ar = 1 THEN 30 ELSE 10 END) as st1,\
                       (CASE WHEN transfer = 1 THEN 50 ELSE 10 END) as st2, \
                       ST_Distance(geography(ST_MakePoint(longitude, latitude)),  \
                          geography(ST_MakePoint('{job[1]}', '{job[0]}')))/1000 as km \
                        FROM {table} \
                        WHERE rent='{input_rent}' and deposit <= '{input_deposit}' \
                        and pay <= '{input_pay}' and recent = 1) as r) \
                    SELECT address, rent, size, deposit, pay, latitude, longitude, criteria, km/{gu} as distance,\
                    ((st1 + st2)/20 + time/10) as score \
                    From (SELECT *, \
                            (CASE WHEN km/{gu} <= 10 THEN 100 \
                                  WHEN km/{gu} between 10 and 90 THEN (100 - km/{gu}) ELSE 0 END) as time\
                          FROM FIRST) as k \
                    ORDER BY score desc LIMIT 10;")        
        
    # 출력된 SELECT 결과를 pandas DataFrame으로 변환
    row = psql.read_sql(sql, conn)    

    # 편의시설 상세정보 출력
    for i in range(num):
        # 각각의 편의시설 table을 불러옴
        new_sql = (f"SELECT criteria, con_name, con_addres, con_latitu, con_longit \
                    FROM {globals()['table_con_{}'.format(i)]};")
        new_row = psql.read_sql(new_sql, conn) # 해당 편의시설 SELECT결과를  pandas DataFrame으로 변환
        # 메인데이터에 편의시설 상세데이터를  LEFT JOIN 실시
        all_df = pd.merge(left=row, right=new_row, how='left', on='criteria', sort=False) 
        globals()['con_detail_{}'.format(i)] = all_df
        print(globals()['con_detail_{}'.format(i)]) # 결합된 데이터 출력

    # 구글지도에 표시될 중간지점을 선택
    maps = folium.Map(location=[37.5642135, 127.0016985], zoom_start=12)
    # 각 매물에 핀을 찍음
    
    for j in range(len(row)): 
        out = "Address : {} ".format(row.iloc[j][0])
        popup = folium.Popup(out, max_width=2650)
        print(out)
        folium.Marker([row.iloc[j][5], row.iloc[j][6]], popup=popup,icon=folium.Icon(icon='cloud')).add_to(maps)
        #             out = "Address : {} ".format(row.iloc[i][0])
#             popup = folium.Popup(out, max_width=2650)
#             folium.Marker([row.iloc[i][3], row.iloc[i][4]], popup=popup,icon=folium.Icon(icon='cloud')).add_to(maps)

    return maps
```

## 조건 입력 관련 처리

```python
def latlon(): # 직장의 위도 경도 계산 및 직장의 구의 속도 도출
    address = input('직장 주소 : ')
    gu = address.split(' ')[1]
    cur.execute(f"SELECT speed FROM SPEED \
            WHERE area='{gu}'") #직장이 속한 구의 대중교통 km/분 평균속도

    gu_speed = cur.fetchall()[0][0] 

    url = "https://dapi.kakao.com/v2/local/search/address.json?&query=" + address
    result = requests.get(urlparse(url).geturl(),
                          headers={"Authorization":"KakaoAK ***"})
    json_obj = result.json()
    global val   # 전역 변수를 지역 범위에서 사용하고 싶을 경우
    for document in json_obj['documents']:
        # 값이 NoneType 일 경우의 오류 방지
        try:
            val = [document['y'], document['x']]
        except:
            val = [0, 0]
    return gu_speed, val # 직장의 위치, 직장의 위도 경도

# 조건 입력 받는 함수
def condition(input_house):
    # 집유형에 따라 해당되는 {table} 도출
    if input_house == '오피스텔':
        table = 'officetels'
    elif input_house == '빌라':
        table = 'villa'
    elif input_house == '원룸/투룸':
        table = 'one_two_room'
        
    input_rent = input("전/월세 유형: ")
    # 전월세 유형에 따라 입력받는 것이 달라짐
    if input_rent =='전세': # 보증금만 입력받음
        input_deposit = input("보증금: ")
        gu, job = latlon()
        input_pay = 0
    elif input_rent == '월세':# 보증금과 월세 둘다 입력받음
        input_deposit = input("보증금: ")
        input_pay = input("월세: ")
        gu, job = latlon()
    return solution(table,input_rent,input_deposit,input_pay,gu,job)
```
